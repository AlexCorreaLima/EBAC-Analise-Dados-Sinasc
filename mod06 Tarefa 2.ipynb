```python
# Imports e leitura do CSV
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Profissão Cientista de Dados M6 - SINASC_RO_2019.csv")
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sns.boxplot(x=df["IDADEMAE"].dropna(), color='skyblue')
plt.title("Boxplot da Idade da Mãe")
plt.xlabel("Idade da Mãe")
plt.show()

```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(x=df["PESO"].dropna(), color='lightgreen')
plt.title("Boxplot do Peso do Bebê")
plt.xlabel("Peso (g)")
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
sns.histplot(df["IDADEMAE"].dropna(), bins=20, kde=True, color='skyblue')
plt.title("Histograma da Idade da Mãe")
plt.xlabel("Idade da Mãe")
plt.ylabel("Frequência")
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
import matplotlib.pyplot as plt

# Exemplo de dados de peso do bebê em kg
pesos = [3.2, 3.5, 3.0, 2.8, 3.6, 3.1, 2.9, 3.3, 3.7, 3.4, 3.0, 3.2]

plt.hist(pesos, bins=5, edgecolor='black')
plt.title('Histograma do Peso do Bebê')
plt.xlabel('Peso (kg)')
plt.ylabel('Frequência')
plt.show()
```
